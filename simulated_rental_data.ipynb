```python
# Step 1: We must download import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Step 2: Load the dataset

file_path = 'simulated_rental_data.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataset to understand its structure
df.head()
```

```python
# Step 3: Data Cleaning

# Here: Checking for missing values
missing_data = df.isnull().sum()

# Here: Checking for duplicates
duplicate_rows = df[df.duplicated()]

# Display the number of missing values and duplicate rows
print(f'Missing Values: \n{missing_data}')
print(f'Duplicate Rows: \n{duplicate_rows}')

# If duplicates are found, they can be dropped
df.drop_duplicates(inplace=True)

# If there are missing values, handle them (filling with mean/median or removing)
df.fillna(df.mean(), inplace=True)  # For numerical columns

# After cleaning, display the cleaned dataset
df.head()
```

```python
# Step 4: Exploratory Data Analysis (EDA)

# This part you can see summary Statistics
summary_stats = df.describe()

# This part you can see display summary statistics
print(f"Summary Statistics: \n{summary_stats}")
```

```python
# Step 5: Analyzing distribution of 'Age' column (to see how age is distributed)
plt.figure(figsize=(10,6))
sns.histplot(df['Age'], kde=True, color='blue', bins=30)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 6: Analyzing the frequency of rented car models
plt.figure(figsize=(10,6))
sns.countplot(data=df, y='Rented Car', order=df['Rented Car'].value_counts().index, palette='viridis')
plt.title('Distribution of Rented Car Models')
plt.xlabel('Count')
plt.ylabel('Car Model')
plt.show()
```

```python
# Step 7: Relationship between 'Distance Travelled' and 'Amount Paid'
df['Distance Travelled Binned'] = pd.cut(df['Distance Travelled'], bins=10)

plt.figure(figsize=(130,50)) # We increase the axis area
sns.countplot(data=df, x='Distance Travelled Binned', hue='Rented Car', palette='Set1', width=0.8, dodge=True)  

# We increase the thickness of the bar with width
# dodge=True separates the bars

plt.title('Count of Distance Travelled by Customer and Rented Car Status', fontsize=130)
plt.xlabel('Distance Travelled (km)', fontsize=100)
plt.ylabel('Count', fontsize=100)

# We adjusted the spacing of the labels on the Y axis
plt.yticks(fontsize=80,rotation=0, ha='right')

#We adjusted the spacing of the labels on the X axis
plt.xticks(fontsize=70, rotation=50, ha='right')

# We made it increase the size of the legend
plt.legend(title='Rented Car', title_fontsize='16', fontsize=90)

plt.show()
```

```python
# Step 8: Analyze Car Brand Preferences by Age Group
age_bins = [18, 25, 35, 45, 55, 65, 100]  # Here we define age ranges
age_labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']  # Here we define labels
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Here the group data by age group and rented car, then count preferences
age_brand_preferences = df.groupby(['Age Group', 'Rented Car']).size().reset_index(name='Count')

# We adjusted the pivot table for better visualisation
pivot_table = age_brand_preferences.pivot(index='Age Group', columns='Rented Car', values='Count').fillna(0)


plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', fmt='.0f', cbar=True)
plt.title('Car Brand Preferences by Age Group', fontsize=16)
plt.xlabel('Car Brand', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()
```

```python
# Step 9:Group by Age Group and Rented Car to count preferences
age_car_preference = df.groupby(['Age Group', 'Rented Car']).size().reset_index(name='Count')

# Print all preferences sorted by Age Group and Count
print("Car preferences by age group:")
print(age_car_preference.sort_values(by=['Age Group', 'Count'], ascending=[True, False]))

# This part we find the most active age group
most_active_age_group = df['Age Group'].value_counts().idxmax()

# Than we filter for the most active age group
most_active_age_group_cars = age_car_preference[age_car_preference['Age Group'] == most_active_age_group]

# After find the most rented car in the most active age group
most_rented_car_in_age_group = most_active_age_group_cars.sort_values(by='Count', ascending=False).iloc[0]

# Lastly print the result for the most active age group
print(f"\nThe most active age group is '{most_active_age_group}', and their most rented car is '{most_rented_car_in_age_group['Rented Car']}' with {most_rented_car_in_age_group['Count']} rentals.")

```

```python
# Step 10:Group by Age Group to count total rentals
age_rentals = df['Age Group'].value_counts().reset_index()
age_rentals.columns = ['Age Group', 'Rental Count']

# Print the most active age group
print("Total rentals by age group:")
print(age_rentals.sort_values(by='Rental Count', ascending=False))
```

```python
# Step 11:Count the rentals for each car
most_rented_car = df['Rented Car'].value_counts().reset_index()
most_rented_car.columns = ['Car Model', 'Rental Count']

# Here print the most rented car
print("Most rented cars:")
print(most_rented_car)

# Than get the top car
print(f"\nThe most rented car is: {most_rented_car.iloc[0]['Car Model']} with {most_rented_car.iloc[0]['Rental Count']} rentals.")
```
